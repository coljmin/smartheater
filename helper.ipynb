```python
import json


stamp = 1578438000
count = 0

with open('winter_day.json', 'r') as openfile:
    weather = json.load(openfile)


for i in range(96):
    weather['2020-01-08'][str(i)]['timestamp'] = stamp
    stamp += 900
    count += 1


print(weather)
print(count)

with open("winter_day.json", "w") as outfile:
    json.dump(weather, outfile)
```

```python
y = []
x = []

with open('winter_day.json', 'r') as openfile:
    weather = json.load(openfile)


for i in range(96):
    temp = weather['2020-01-08'][str(i)]['temperature']
    time = weather['2020-01-08'][str(i)]['timestamp']
    y.append(temp)
    x.append(time)
```

```python
import seaborn as sns

sns.lineplot(x=x, y=y)
```
